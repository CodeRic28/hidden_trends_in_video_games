````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

sns.set_style('darkgrid')
````

# Reading Dataset
First off we will read the dataset.For this we use the Pandas library
Since the Pandas read.csv function automatically adds an index column, we will replace it and set our own index column with the column that's already present in the dataframe

````python
vg_data = pd.read_csv('./discovering-hidden-trends-in-global-video-games/Video Games Sales.csv')
vg_data = vg_data.set_index('index')
vg_data
````

## Shape of Dataset
Shape of the dataset basically means the dimensions of the dataset. It includes the total number of rows and columns that the dataset contained

````python
# Shape of the dataset
vg_data.shape
````

````python
print(f'The dataset contains {vg_data.shape[0]} rows and {vg_data.shape[1]} columns')
````

````python
# Checking for null values
null_data = vg_data[vg_data.isnull().any(axis=1)]
null_data
````

Here we've isnull() function to check if the dataset contains any null values. Null values are the values that are empty or not available. For instance NaN stands for Not a Number and it fills in the space if a column containing integer or float values has empty entries.
Further, after finding the null values, we can later manipulate them as we like. We can either delete the rows containing null values or fill in the values.

````python
schema = pd.read_csv('./discovering-hidden-trends-in-global-video-games/column_schema.csv', index_col='Column').Description
schema
````

In the above code, we've read a new file. This file contains 2 columns - first with the column names in the original (vg_data) dataset and the other containing the description of each column.
This can come in handy when plotting graphs to actually find out what each column defines.

````python
# Checking for numeric values
vg_data.info()
````

the info() function is used to gather information regarding a dataset. Here, we can clearly see the names of the columns, if the columns have null values, and each column's data type.
By looking at the datatypes of each column, it is clear that how many of the columns contain numbers - the datatype that we're most interested in as it helps in plotting the values
It's clear that 8 out of 12 columns have a number datatype - either integer or float

````python
vg_data.describe()
````

The describe() function comes even more in handy. It not only tells us the name of the columns but also some of the major statistical operations. For instance, the first row shows the count of each column, second row contains the averages of each column, third tells how much the data is deviated from average and so on.
Note that the describe() function only picks out columns that contain numbers

## Total number of Games in each Genre
There are over 1900 games in the dataset but not that many genres. It is clear that many games come under the same genre. Next we will see exactly how many games there are in each genre

````python
# Find total number of games according to Genre
plt.figure(figsize=(12,5))
plt.title("Total Games in each Genre", fontsize='15')
genre = vg_data.Genre.unique()
count_by_genre = vg_data.groupby(['Genre'])['Genre'].count()

plt.xticks(ticks=np.arange(12), labels=genre, rotation=20)
barPlotGenre = sns.barplot(x=genre, y=count_by_genre)
barPlotGenre = barPlotGenre.set(xlabel="Genre", ylabel="Total number of Games")
plt.show()
````

By plotting the graph we can see how many games are there in each genre. The x-axis contains the names of the genres and y-axis contain the number of games.
We can see that the genre with maximum number of games is Adventure, and at the second place is Sports.
The genre with least number of games is Strategy. Apparently, not a lot of people like to play strategic games.

````python
# How many of the Games are published by Nintendo
publisher = vg_data.Publisher.nunique()
print(publisher)
````

Here, we've found the total number of games published by a publisher - in this case Nintendo.
If we use the nunique() function, it will give us distinct names under the publisher's name

let's sort and arrange some data into variables that we can later use:
Here, we're taking the top 10 values from the columns listed below and sorting them in descending order

````python
north_sales_df = vg_data[['Game Title', 'North America']].sort_values('North America', ascending=False).head(10)
europe_sales_df = vg_data[['Game Title', 'Europe']].sort_values('Europe', ascending=False).head(10)
global_sales_df = vg_data[['Game Title', 'Global']].sort_values('Global', ascending=False).head(10)
europe_sales_df
````

For instance, the table above contains the Title of the games that are sold the most in Europe. Let's plot line graphs to see what game is the highest seller and what game is the lowest.

````python
# Top 10 most sold games iN North America
plt.figure(figsize=(12,5))
plt.title(schema['North America'], fontsize='15')
sns.lineplot(x=north_sales_df['Game Title'], y=north_sales_df['North America'])
plt.xticks(rotation=30)
plt.show()
````

We can see that the most sold game in North America is Wii Sports and the least sold is Super Mario World. Super Mario Bros on the other hand is the 2nd most sold game in NA.Seems like people like the good ol'classic version of Super Mario

````python
# Top 10 most sold games in Europe
plt.figure(figsize=(15,5))
plt.title(schema['Europe'], fontsize='15')
sns.lineplot(x=europe_sales_df['Game Title'], y=europe_sales_df['Europe'], marker='o')
plt.xticks(rotation=30)
plt.show()
````

Observing the graph of Europe, Wii Sports again takes the lead as the most sold game. This time, however, the least sold game is Wii Fit Plus.
The sales of Wii Sports is significantly more than any other game with 27.6 million sales as compared to the 2nd most sold game with just 12.5 copies sold.
Wii Sports is a beast when it comes to gaming. After the top 5 games, the number of sales of the remaining games does not differ by much

````python
# Top 10 most sold games Globally
plt.figure(figsize=(12,5))
plt.title(schema['Global'], fontsize='15')
sns.lineplot(x=global_sales_df['Game Title'], y=global_sales_df['Global'], marker='X')
plt.xticks(rotation=30)
plt.show()
````

Not only NA and EU, but all over the world, people absolutely love Wii Sports that has over 80 million copies sold and the second best is Super Mario Bros. with just 40 million copies sold. For Wii Sports, it's not even a competition. The worst game globally is Nintendogs with less than 25 million copies sold. The rest of the games are doing average. Wii Sports is the outlier and since the extreme value is very large, this will have a huge impact on the average sales.

````python
# Top 10 most sold games combined
plt.figure(figsize=(16,5))
plt.title("Total sales in America, Europe and Globally comparison", fontsize='15')
sns.lineplot(x=north_sales_df['Game Title'], y=north_sales_df['North America'], marker='^')
sns.lineplot(x=europe_sales_df['Game Title'], y=europe_sales_df['Europe'], marker='o')
sns.lineplot(x=global_sales_df['Game Title'], y=global_sales_df['Global'], marker='X')
plt.xticks(rotation=50)
plt.show()
````

Here, we've just plotted all the sales figures of NA, EU and the rest of the world, in one figure to get a better picture of the sales. In all the three graphs, Wii Sports is the best with second best Super Mario Bros. Duck hunt, another classic coming third with about 30 million sales.

## Finding the average sales in North America, Europe and the rest of the world

````python
plt.figure(figsize=(12,5))
plt.title('Averages', fontsize='15')
north_mean = vg_data.iloc[:,6].mean()
europe_mean = vg_data.iloc[:,7].mean()
japan_mean = vg_data.iloc[:,8].mean()
rest_mean = vg_data.iloc[:,9].mean()
data=[north_mean, europe_mean, japan_mean, rest_mean]
labels=['North America', 'Europe', 'Japan', 'Rest of the World']
plt.pie(data, labels=labels, explode=[.01,.05,.05,.07], autopct='%1.2f%%')

plt.show()
````

Here we've plot the average sales in NA, EU and Japan. Recall from previous sales graphs that Wii Sports sold the most copies in all of the continents so that will have an impact on the averages.
From the figure above we can see that North America has the most sales accounting to 50% of the total copies sold.
2nd comes Europe with 28.39% of the total copies sold followed by Japan with 12.75%. The rest of the world is just 8.29%. I think it is safe to say that people in NA play the most games and spend the most amount of money on games too.
